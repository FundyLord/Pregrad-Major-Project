#Balancing imbalanced data using SMOTE

```python
from imblearn.over_sampling import SMOTE

# Separate features (X) and target variable (y)
X = df.drop('Loan_Status_label', axis=1)
y = df['Loan_Status_label']

# Apply SMOTE to oversample the minority class
smote = SMOTE(random_state=42)  # You can adjust the random_state
X_resampled, y_resampled = smote.fit_resample(X, y)

# Create a new DataFrame with the resampled data
df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
df_resampled['Loan_Status_label'] = y_resampled

# Print the value counts of the target variable in the resampled data
print(df_resampled['Loan_Status_label'].value_counts())
```

#Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
log_x = df_resampled.drop('Loan_Status_label', axis=1)
log_y = df_resampled['Loan_Status_label']
```

```python
log_x
```

```python
log_y
```

```python
df.shape
```

```python
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(log_x, log_y, test_size=0.2, random_state=42)

# Initialize and train the Logistic Regression model
logreg = LogisticRegression()
logreg.fit(x_train, y_train)

# Make predictions on the test set
log_y_pred = logreg.predict(x_test)

# Evaluate the model
accuracy = accuracy_score(y_test, log_y_pred)
print(f"Accuracy: {accuracy}")

# Generate and print the classification report
print(classification_report(y_test, log_y_pred))

# Generate and display the confusion matrix
conf_matrix = confusion_matrix(y_test, log_y_pred)
print("Confusion Matrix:")
conf_matrix
```

```python
#Training and test prediction
log_y_pred_train = logreg.predict(x_train)
log_y_pred_test = logreg.predict(x_test)
```

# Model Comparison based on Test and Training Accuracy


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

metrics_dict = {
    'Model': [],
    'Train Accuracy': [],
    'Test Accuracy': [],
}

for model_name, train_accuracy, test_accuracy in [
    ('Logistic Regression', log_train_accuracy, log_test_accuracy),
    ('SVM', svm_train_accuracy, svm_test_accuracy),
    ('KNN', knn_train_accuracy, knn_test_accuracy),
    ('Decision Tree (Post-Pruning)', post_dtree_train_accuracy, post_dtree_test_accuracy),
    ('Decision Tree (Pre-Pruning)', pre_dtree_train_accuracy, pre_dtree_test_accuracy)
]:
    metrics_dict['Model'].append(model_name)
    metrics_dict['Train Accuracy'].append(train_accuracy)
    metrics_dict['Test Accuracy'].append(test_accuracy)

metrics_df = pd.DataFrame(metrics_dict)
metrics_df
```

```python
#Function to make Loan aprroval prediction
def predict_loan_status(logreg, pdays, previous, no_previous_contact, poutcome_success):

    # Create a DataFrame with user input
    user_input = pd.DataFrame({
        'pdays': [pdays],
        'previous': [previous],
        'no_previous_contact': [no_previous_contact],
        'poutcome_success': [poutcome_success]
    })

    # Make prediction
    prediction = logreg.predict(user_input)[0]

    # Map prediction to loan status label
    if prediction == 1:
      loan_status = 'Loan Approved'
    else:
      loan_status = 'Loan Rejected'

    return loan_status

# Getting user input
pdays = int(input("Enter number of days since the client was last contacted('0' means never contacted): "))
previous = int(input("Enter number of contacts performed before this campaign for the client: "))
no_previous_contact = int(input("Enter whether client did have previous contacts or not('1': no contact or '0': had previous contacts): "))
poutcome_success = int(input("Enter previous campaign outcome('1':success or '0':no success): "))

# Printing prediction
loan_status = predict_loan_status(logreg, pdays, previous, no_previous_contact, poutcome_success)
print(f"Predicted Loan Status: {loan_status}")
```
